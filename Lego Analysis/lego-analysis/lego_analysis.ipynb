# 1.What percentage of all licensed sets ever released were Star Wars themed?

#### Join 'parent_theme' from lego_sets.csv and 'name' from parent_theme.csv:
Syntax: left_df.merge(right_df, left_on = 'left_column', right_on = 'right_column')

```python
merged_df = df.merge(parent_themes, left_on = 'parent_theme', right_on = 'name')
merged_df.head()
```

On doing merge(), we got 2 columns: name_x, name_y. They have the same values, so we can drop name_y.

```python
merged_df.drop(columns = 'name_y', inplace = True)
## Accidentally I had to run this cell again, so it threw an error.
## A column can't be dropped twice.
```

```python
merged_df.head()
```

Get the rows where is_licensed is True

```python
licensed = merged_df[merged_df['is_licensed']] ## is_licensed is bool.
licensed.head()
```

Get the rows where parent_theme is Star Wars

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head()
```

```python
star_wars.shape ## Check length of star_wars df.
```

```python
star_wars.shape[0]
```

```python
licensed.shape[0]
```

Get the PERCENTAGE of all LICENSED sets released with parent theme as Star Wars.

```python
the_force = int(star_wars.shape[0]/licensed.shape[0] * 100)
print(the_force)
```

## This value is incorrect, since the licensed df contains some NaN values which need to be removed from the licensed df.

NaN values in other columns are acceptable. If we have NaN in 'set_num', it means its a duplicate product, which SHOULD be dropped. That is why, subset = ['set_num'] is mentioned.

```python
merged_df[merged_df['set_num'].isnull()].shape
```

```python
licensed = licensed.dropna(subset=['set_num'])
```

```python
the_force = int(star_wars.shape[0]/licensed.shape[0] * 100)
print(the_force)
```

# 2.In which year was Star Wars not the most popular licensed theme?

```python
licensed.head()
```

We do groupby on licensed df, based on 'year' and 'parent_theme'.

```python
licensed_sorted = licensed.sort_values('year')
# licensed_sorted.groupby(['year','parent_theme']).head()
```

We are checking HOW MANY Star Wars, Batman, Disney, etc are there in each year. We use sum() for this aggregation purpose.

```python
licensed_sorted['count'] = 1

licensed_sorted.groupby(['year','parent_theme']).sum().head()
```

Reset the index after doing aggregation - ?

```python
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

We want to get only the max row anything that shares the SAME YEAR. We use drop_duplicates(subset=['year']) to drop duplicates based on year.

```python
max_df = summed_df.sort_values('count', ascending = False).drop_duplicates(subset=['year'])
max_df.head()
```

```python
max_df.sort_values('year', inplace = True)
max_df
```

As we see above, Star Wars was popular until 2017, when Super Heroes took over.
